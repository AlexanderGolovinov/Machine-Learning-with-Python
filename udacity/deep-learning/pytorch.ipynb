```python
# torch – This is the core PyTorch library used for building and training models.
# torch.nn – Contains classes to create neural network layers and activation functions.
import torch
import torch.nn as nn

class MLP(nn.Module):
        def __init__(self, input_size):
            super(MLP, self).__init__()
            self.hidden_layer = nn.Linear(input_size, 64)
            self.output_layer = nn.Linear(64, 2)
            self.activation = nn.ReLU()

        def forward(self, x):
            x = self.activation(self.hidden_layer(x))
            return self.output_layer(x)

model = MLP(input_size=10)
print(model)
# MLP(
#   (hidden_layer): Linear(in_features=10, out_features=64, bias=True)
#   (output_layer): Linear(in_features=64, out_features=2, bias=True)
#   (activation): ReLU()
# )

model.forward(torch.rand(10))
# tensor([0.2294, 0.2650], grad_fn=<AddBackward0>)
```

```python
import matplotlib.pyplot as plt

# Define the MLP model
class MLP(nn.Module):
    def __init__(self, input_size):
        super(MLP, self).__init__()
        self.hidden_layer = nn.Linear(input_size, 64)
        self.output_layer = nn.Linear(64, 2)
        self.activation = nn.ReLU()

    def forward(self, x):
        x = self.activation(self.hidden_layer(x))
        return self.output_layer(x)

# Initialize the model and input data
model = MLP(input_size=10)
input_data = torch.rand(10)

# Visualize the data flow
hidden_output = model.activation(model.hidden_layer(input_data))
final_output = model.output_layer(hidden_output)

# Plotting
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.title("Hidden Layer Output")
plt.bar(range(64), hidden_output.detach().numpy())

plt.subplot(1, 2, 2)
plt.title("Final Output (2 neurons)")
plt.bar(range(2), final_output.detach().numpy())

plt.show()
```
